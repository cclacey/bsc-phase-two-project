```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def main():
    fig, ax = plt.subplots()

    

    sns.set_style("darkgrid")
    ax = sns.scatterplot(x='lat',y='long',data=df,hue='price')
    sns.scatterplot(x='lat', y='long', data=df,marker='o',edgecolor="blue",s=70,ax=ax)
    
    x = df_WF['latitude']
    y = df_WF['longitude']
    image_path = get_sample_data('C:/Users/umbac/Documents/Flatiron/Phase_2/project/bsc-phase-two-project/notebooks/exploratory/Warren/images/Whole-Foods-Logo.png')
    imscatter(x, y, image_path, zoom=.06, ax=ax)
    ax.plot(x, y)
    plt.show()
    
    plt.savefig('..\..\..\images\LatLongWF.jpg')

    
    
    
def imscatter(x, y, image, ax=ax, zoom=0.06):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
main()
```
